```python
def replace_with_nan_in_column(dataframe, column_name, values_to_replace):
    def replace_values(cell):
        if pd.isna(cell):
            return cell
        parts = [part.strip() for part in str(cell).split(',')]
        for part in parts:
            if part in values_to_replace:
                return np.nan
        return cell

    dataframe[column_name] = dataframe[column_name].apply(replace_values)
    return dataframe
```

### Seperate the Flavors for Eda Visuals

```python
categories = {
    'sweet': [
        'agave nectar', 'almond milk', 'amaretto', 'apple schnapps', 'butterscotch schnapps',
        'cinnamon schnapps', 'honey', 'maple syrup', 'simple syrup', 'sugar', 'vanilla syrup'
    ],
    'citrus': [
        'lemon juice', 'lime juice', 'orange juice', 'grapefruit juice', 'pineapple juice', 'blood orange juice'
    ],
    'bitter': [
        'absinthe', 'amaro', 'aperol', 'campari', 'angostura bitters', 'aromatic bitters',
        'tonic water', 'vermouth', 'yellow chartreuse'
    ],
    'herbal': [
        'basil', 'cilantro', 'mint', 'rosemary', 'thyme'
    ],
    'spicy': [
        'ginger', 'cinnamon', 'cayenne pepper'
    ]
}
```

```python
def categorize_drink(ingredients,categories): 
    tags = {key: False for key in categories.keys()}
    for ingredient in ingredients:
        for category, items in categories.items():
            if ingredient in items:
                tags[category] = True
    return tags
```

```python
for category in categories.keys():
    clean_cocktail[category] = False
```

```python
for index, row in clean_cocktail.iterrows():
    drink_ingredients = row[[col for col in clean_cocktail.columns if 'ingredient' in col]].dropna().tolist()
    tags = categorize_drink (drink_ingredients, categories)
    for category, tagged in tags.items():
        clean_cocktail.at[index, category] = tagged
```

```python
### handles vectors lets say the input vector size is smaller than training vector or empty this makes sure we can do consine similarity
def get_recipe_vector(tokens, model, vector_size):
    word_vectors = [model.wv[word] for word in tokens if word in model.wv]
    if word_vectors:
        mean_vector = np.mean(word_vectors, axis=0)
        if mean_vector.shape[0] == vector_size:
            return mean_vector
        elif mean_vector.shape[0] > vector_size:
            return mean_vector[:vector_size]
        else:
            return np.pad(mean_vector, (0, vector_size - mean_vector.shape[0]), 'constant')
    else:
        return np.zeros(vector_size)
```
